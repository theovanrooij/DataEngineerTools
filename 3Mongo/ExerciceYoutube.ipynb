## Question 0

### Netoyer les données

On veut récupérer : id - title

```python
import json
with open('./data/US_category_id.json') as f:
    categories_json = json.load(f)
```

```python
categories_items = categories_json["items"]
categories = []
for category in categories_items:
    category_dict = dict()
    category_dict["id"] = category["id"]
    category_dict["title_category"] = category["snippet"]["title"]
    categories.append(category_dict)
df_categories = pd.DataFrame.from_dict(categories)
df_categories["id"] = df_categories["id"].astype("int64")
```

```python
df_youtube = df_youtube.merge(df_categories, left_on='category_id', right_on='id')
```

```python
df_youtube["channel_title"] = df_youtube["channel_title"].astype("str")
```

```python
df_youtube[df_youtube["channel_title"] == "Apple"]
```

```python
dict_youtube = df_youtube.to_dict('index')
```

## Question 6 

Récupérer les chaines Youtube avec la plus grande moyenne de likes.

```python
cur = collection.aggregate([
    {"$group" : {"_id" : "$channel_title", "Average Like" : {"$avg" : "$likes"}}},
    {"$sort" : {"Average Like" :-1}}
])
list(cur)
```
